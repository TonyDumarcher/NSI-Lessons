# Récursivité 1  
En classe de première, les styles de programmation étudiés sont :
1.	Le style impératif, les séquences d’instructions sont exécutées l’une après l’autre ;
2.	Le style itératif, des instructions sont exécutés dans des boucles while ou for.

Une fonction récursive est une fonction qui s’appelle elle-même ce qui va permettre, par exemple, de supprimer une boucle for.


**Exemple de style itératif avec la boucle ```for```**  
Dans le code suivant, la variable i prend successivement les valeurs entières de 0 à 9 qui sont ensuite affichés.

````python
for i in range(10):
    print(i)
````

A la fin de ce code, la variable i existe et a la valeur 9.

**Transformation en fonction récursive**  
La fonction <i> affiche </i> est une fonction récursive : plus de for (ni de while), pas d'instruction d'affectation, aucune variable existe en dehors de la fonction.

````python
def affiche_rec(k):
    if k < 10:
        print(k)
        affiche_rec(k + 1)
````

````python
affiche_rec(0)
````

**Exemple de fonction itérative avec la boucle ```while```** 
Fonction <i> rebours1 </i> affiche un compte à rebours avec une boucle <i> while </i>.

````python
def rebours1(n):
    while n >= 0:
        print(n)
        n = n - 1
````

````python
rebours1(5)
````

**Exemple de fonction itérative avec la boucle ```for```**  
Fonction <i> rebours2 </i> affiche un compte à rebours avec une boucle <i> for </i>.

````python
def rebours2(n):
    for i in range(n+1):
        print(n-i)
````

````python
rebours2(5)
````

**Transformation en fonction récursive**  
Fonction récursive <i> reboursRec </i> affiche un compte à rebours sans boucle <i> for </i> ou <i> while </i>.

````python
def rebours_rec(n):
    if n >= 0:
        print(n)
        rebours_rec(n - 1)
````

````python
rebours_rec(5)
````

**Fonction récursive de dessin** : <i> dessineRec </i>

````python
from turtle import *

def dessine_rec(n):
    if n > 0:
        forward(n)
        right(90)
        dessine_rec(n-5)
````

Faire une telle figure sans utiliser la récursivité est assez complexe, éventuellement difficilement surmontable.  
**Conclusion**  
Le principe de la récursivité est de réduire un problème complexe à un ensemble de problèmes de plus en plus simples qui tendent vers le cas de base (aussi appelé *cas trivial*).

**Fonction récursive** du produit de deux nombres à l'aide de la somme.

````python
def produit_rec(n,p):
    if p == 0:
        return 0
    else:
        return n+ produit_rec(n,p-1)
````

````python
produit_rec(4,3)
````

**Remarques** :
1. Le test p == 0 est une condition d'arrêt indispensable, sinon le programme boucle infiniment.
2. Les valeurs passées en paramètres dans les appels récursifs doivent être différents. Sinon la fonction s'exécute à chaque appel de manière identique et continue de s'exécuter indéfiniment.
3. Après un nombre fini d'appels, la ou les valeurs passées en paramètres doivent permettre de satisfaire la condition d'arrêt.

# **Comparaison itératif/récursif**  

|                  Itératif                        |           Récursif               | 
| :----------------------------------------------: |: ------------------------------: | 
| Fonctionne à l'aide de boucles for ou while      | En général, pas de boucle for ou while.|
| Onrépète (itère) $n$ fois la même opératio       | On appelle la même opération, mais à chaque appel, les paramètres changent et l'opération est plus simple. Cela s'appelle le "cas général".   |  
| Le paramètre qui varie est l'incrément de boucle.| Le paramètre qui varie est le (ou les) paramètre qui est passé en argument du sous-programme qui s'appelle lui-même.  |  
| La condition de sortie de boucle détermine l'arrêt de la boucle.| Il y a toujours un cas (souvent introduit par un if), dit "cas de base", qui permet de mettre fin aux appels récursifs. Chaque appel récursif doit permettre de se rapprocher du cas de base.  |
| Le problème est résolu lorsque la boucle est quittée.| Le problème est résolu lorsque l'on atteint le cas de base.  |


**Concevoir un programme récursif**
1. Trouver le cas général, soit chercher une décomposition récursive du problème.
2. Trouver le cas de base, soit Trouver la condition d'arrêt de la récursivité et la solution du problème dans ce cas particulier.

**Exercice 1**  
Les seules opérations autorisées sur les nombres entiers sont soit ajouter 1, soit retrancher 1. La fonction somme renvoie la somme des entiers positifs a et b.

````python
def somme(a,b):
    while b>0:
        a = a + 1
        b = b - 1
    return a
````

````python
somme(6, 3)
````

Ecrire une version récursive de cette fonction.

````python
def somme_rec(a, b):
    if b <= 0:
        return a
    else:
        return somme_rec(a+1, b-1)
````

````python
somme_rec(3, 6)
````

**Exercice 2**  
En s’inspirant de l’exercice précédent, écrire une fonction qui renvoie la somme de deux nombres entiers quelconques. Les seules opérations autorisées sont soit ajouter 1, soit retrancher 1.
1. Ecrire une fonction sommes itérative avec une boucle while.
2. Ecrire une fonction sommes_rec récursive.

````python
def sommes(a,b):
    while b:
        if b>0:
            a = a + 1
            b = b - 1
        else:
            a = a - 1
            b = b + 1
    return a
````

````python
sommes(6, -3)
````

````python
def sommes_rec(a, b):
    if b == 0:
        return a
    else:
        if b > 0:
            return sommes_rec(a+1, b-1)
        else:
            return sommes_rec(a-1, b+1)
````

````python
sommes_rec(6, -3)
````

Soit la fonction permettant de savoir si un entier naturel est pair ou non.

````python
def pair(n):
    while n>0:
        n = n - 2
    return n == 0
````

````python
def pair_rec(n):
    if n <= 0:
        return n == 0
    else:
        return pair_rec(n-2)
````

Transformer votre fonction pour tout entier relatif.

````python
def pair_rec_relatif(n):
    if abs(n) <= 1:
        return n == 0
    else:
        return pair_rec_relatif(abs(n)-2)
````
