## 2ème exemple : suite de Fibonacci

<span style="color:red"> **Définition de la Suite de Fibonacci**
    
On appelle (f<sub>n</sub>) la suite de Fibonacci telle que, 
- f<sub>0</sub>=0 et f<sub>1</sub>=1 ;
- f<sub>n</sub>=f<sub>n-1</sub>+f<sub>n-2</sub>, quelque soit n entier naturel, différent de 0 et de 1.


Calculer les onze premiers termes de la suite de Fibonacci.

**Ecrire une fonction <i>fibo1</i>, d'argument n, un  entier naturel, avec la boucle <i> while</i> qui renvoie le terme de la suite de rang $n$.**

```python
def fibo1(n):
    u, v, compteur = 0, 1, 0 # u est un terme de la suite et v, le terme suivant
    while compteur < n:
        compteur += 1
        u, v = v, u+v
    return u
```

**Test**

```python
fibo1(10)
```

**Ecrire une fonction <i>fibo2</i>, d'argument n, un entier naturel, avec la boucle <i> for.**

```python
def fibo2(n):
    u, v = 0, 1
    for i in range(n):
        u, v = v, u+v
    return u
```

**Test**

```python
fibo2(10)
```

**Ecrire une fonction récursive profonde <i> fiboRec </i>, d'argument n, un entier naturel**

```python
def fibo_rec(n):
    if n==0 or n==1:
        return n
    else:
        return fibo_rec(n-2) + fibo_rec(n-1)
```

**Test**

```python
fibo_rec(10)
```

```python
fibo_rec(30)
```

**Test de rapidité d'exécution**

```python
for i in range(40):
    print(i,fibo_rec(i))
```

Qu'observez-vous? Quelle explication peut-on donner?

**Ecrire une fonction récursive terminale <i> fibo_terminale </i> avec accumulateur.**

```python
def fibo_terminale(n, a=0, b=1):
    if n == 0:
        return a
    else:
        return fibo_terminale(n-1, a+b, a)
```

**Test**

```python
fibo_terminale(30)
```

Comparer les temps d'exécution avec les instructions suivantes :

```python
from timeit import default_timer as timer
debut=timer()
print(fibo_terminale(39))
fin = timer()
temps = fin - debut
print(temps)
```

On peut obtenir une écriture moins simple, mais plus intuitive en insérant un compteur cpt.

**Ecrire une fonction récursive terminale <i> fibo_terminale2 </i> avec un compteur.**

```python
def fibo_terminale2(n, cpt=0, a=0, b=1):
    if cpt == n:
        return a
    else:
        return fibo_terminale2(n, cpt + 1, a+b, a)
```

```python
fibo_terminale2(30)
```

```python
from timeit import default_timer as timer
debut=timer()
print(fibo_terminale2(39))
fin = timer()
temps = fin - debut
print(temps)
```
