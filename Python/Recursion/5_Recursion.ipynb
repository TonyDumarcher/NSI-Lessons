**Exercice n°5**	Programmation le tri par inversion de manière récursive  
a.	On rappelle l’algorithme du tri par inversion vu en première, en pseudo-langage.
```
fonction trier_par_insertion(tableau):
    n := longueur(tableau)
    Pour i allant de 1 à n-1 faire
        x := tableau[i]
        j := i
        Tant que ((j>0) et (x<tableau[j-1])) faire
            tableau[j] := tableau[j-1]
            j := j-1
            FinTantQue
        tableau[j] := x
        FinPour
```
Traduire en Python et tester un programme.

````python
def trier_par_insertion(tab):
    n = len(tab)
    for i in range(1, n):
        x = tab[i]
        j = i
        while j > 0 and x < tab[j-1]:
            tab[j] = tab[j-1]
            j = j - 1
        tab[j] = x
    return tab
````

````python
trier_par_insertion([1, 11, 6, 89, 10])
````

b. Décomposer cette fonction en deux fonctions : la fonction trier(tableau) qui fait appel à la fonction insertion ayant en paramètres une liste, un élément de la liste et son indice.

````python
def insertion(tab, x, i):
    j = i
    while j > 0 and x < tab[j-1]:
        tab[j] = tab[j-1]
        j = j - 1
    tab[j] = x


def trier(tab):
    n = len(tab)
    for i in range(1, n):
        x = tab[i]
        insertion(tab, x, i)
    return tab
````

````python
trier([1, 11, 6, 89, 10])
````

c. 	Ecrire la fonction récursive insertion_rec qui prend trois paramètres : une liste, un élément de la liste et son indice. Si l’indice est strictement positif, on suppose les éléments d’indice 0 à indice-1 triés et la fonction insère l’élément à la bonne place. 

````python
def insertion_rec(tab, x, i):
    if i == 0 or x >= tab[i - 1]:
        tab[i] = x
        return tab
    tab[i] = tab[i - 1]
    return insertion_rec(tab, x , i - 1)
````

````python
print(insertion_rec([1, 11, 6, 89, 10], 5, 3))
````

d. Ecrire une fonction récursive trier_par_insertion_rec qui prend en paramètres : une liste et la longueur de la liste et qui trie la liste.



````python
def trier_par_insertion_rec(tab, n):
    if n <= 1:
        return tab
    trier_par_insertion_rec(tab, n - 1)
    x = tab[n - 1]
    i = n - 1
    insertion_rec(tab, x, i)
    return tab
````

````python
print(trier_par_insertion_rec([1, 11, 6, 89, 10], 5))
````

**Exercice n°6**	La valeur du pgcd (plus grand commun diviseur) de deux entiers naturels a et b est calculable à l’aide de l’algorithme d’Euclide :  
$pgcd(a,b)=pgcd(b,r)$ où r est le reste de la division euclidienne de a par b.  
En voici un algorithme :  
```
fonction pgcd(a, b):
    si b est nul alors
        retourner a
    Tant que b est différent de 0 faire
        a = b
        b = reste de la division euclidienne de a par b
        FinTantQue
    retourner a
```
Ecrire une version récursive de cette fonction.

````python
def pgcd_recursif(a, b):
    if b == 0:
        return a
    else:
        reste = a % b
        return pgcd_recursif(b, reste)
````

````python
pgcd_recursif(1071, 462)
````

**Exercice n°7**		Ecrire une fonction récursive ```inverse_rec``` qui prend en paramètres une chaîne de caractères ```ch``` et renvoie la chaîne obtenue en inversant l’ordre des caractères.
Par exemple,```inverse_rec("azerty")``` a pour valeur la chaîne ```"ytreza"```.


````python
# Fonction inverse non récursive
def inverse(ch):
    chaine = ""
    for char in ch:
        chaine = char + chaine
    return chaine

# Fonction inverse récursive
def inverse_rec(ch):
    if len(ch) == 1:
        return ch
    else:
        return inverse_rec(ch[1:])
````

````python
inverse("azerty")
````
