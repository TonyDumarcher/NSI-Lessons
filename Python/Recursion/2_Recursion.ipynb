## 1er exemple : fonction factorielle

<span style="color:red"> **Définition de la fonction factorielle :**
- appelée factorielle n , noté n! pour n entier naturel
- 0!=1
- n!=n*(n-1)! pour n non nul</span>

**Calculer à la main : 3!, 5!, 8! :**
1. 3! = ...
2. 5! = ...
3. 8! = ...

**Ecrire une fonction <i> factorielle1 </i>, d'argument n, un  entier naturel, avec la boucle <i> while.**

````python
def factorielle1(n):
    if n == 0:
        return 1
    else:
        i = 0
        res = 1
        while i != n:
            i = i + 1
            res = res * i

        return res
````

**Test**

````python
factorielle1(3)
````

````python
factorielle1(5)
````

````python
factorielle1(8)
````

**Ecrire une fonction <i> factorielle2 </i>, d'argument n, un entier naturel, avec la boucle <i> for.**

````python
def factorielle2(n):
    if n == 0:
        return 1
    else:
        res = 1
        for i in range(n):
            res = res * (i+1)
        return res
````

**Test**

````python
factorielle2(3)
````

````python
factorielle2(5)
````

````python
factorielle2(8)
````

**Ecrire une fonction récursive <i> factorielleRec </i>, d'argument n, un entier naturel, sans oublier le test d'arrêt.**

````python
def factorielle_rec(n):
    if n == 0:
        return 1
    else:
        return factorielle_rec(n - 1) * n
````

**Test**

````python
factorielle_rec(3)
````

````python
factorielle_rec(5)
````

````python
factorielle_rec(8)
````

**Approfondissement** : Décrire les appels récursifs de cette fonction.

**Conclusion** : Un programme récursif nécessite de nombreuses allocations d'espace mémoire par rapport à l'itération, mais il est aussi un style de programmation d'abstraction supérieur. Le programmeur décrit comment calculer la solution à partir d'un cas simple. Au lieu de décrire chaque action à réaliser, on décrit ce que l'on veut obtenir. On dit que c'est de la programmation **déclarative**.  
Une solution récursive est souvent plus intuitive, plus concise et plus lisible qu'une solution impérative comme l'itération.

**La récursivité terminale**  
Une fonction récursive est dite terminale si aucun traitement n'est effectué à la remontée d'un appel récursif (sauf le retour d'une valeur).  
L'intérêt de cette programmation est de ne plus stocker de variables intermédiaires.  
Exemple : Ecrire une fonction récursive ```sommeRec(x)``` qui donne la somme des n premiers entiers naturels.

````python
def sommeRec(x):
    if x == 0:
        return 0
    else:
        return x + sommeRec(x - 1)
````

````python
sommeRec(3)
````

````python
sommeRec(5)
````

````python
sommeRec(10)
````

**Avec un accumulateur**  
en sachant que la fonction terminale aura la forme :   
```
def fonction_terminale(...,acc):
    if condition:
        return acc
    else:
        return instructions et appel récursif
```

````python
def sommeRec_acc(x,acc):
    if x == 0:
        return acc
    else:
        return sommeRec_acc(x - 1, acc + x)
````

````python
sommeRec_acc(3, 0)
````

````python
sommeRec_acc(5, 0)
````

````python
sommeRec_acc(10, 0)
````

**Remarque**  
On peut masquer à l'utilisateur l'utilisation de cette fonction terminale en utilisant une fonction auxiliaire. Cela s'appelle un *wrapper*.

````python
def somme(x):
    return sommeRec_acc(x,0)
````

**Ecrire une fonction récursive terminale <i> factorielle_terminale </i> avec accumulateur.**

````python
def factorielle_terminale(n, acc):
    if n == 0:
        return acc
    else:
        return factorielle_terminale(n - 1, acc * n)
````

**Test**

````python
factorielle_terminale(3,1)
````

````python
factorielle_terminale(5,1)
````

````python
factorielle_terminale(8,1)
````

**Exercice 4** Fonctions puissance  
**Version 1**  
Ecrire une fonction récursive ```puissance_rec1``` qui prend en paramètres un flottant $x$ non nul et un entier naturel $n$ et qui renvoie $x^n$.  
On se base sur la définition mathématique :  
$x^0=1$  
$x^n=x \times x^{n-1}$ pour $n>0$.  
Cette version n’est pas terminale.  

**Version 2**  
Ecrire la version terminale ```puissance_rec1_term```.  

**Version 3**   
Ecrire une fonction récursive ```puissance_rec2``` qui prend en paramètres un flottant $x$ non nul et un entier naturel $n$ et qui renvoie $x^n$.  
On se base sur cette autre méthode :  

$x^0=1$  
$x^n=x^{2k}=(x^k )^2$ pour $n=2k$  
$x^n=x^{2k+1}=x(x^k )^2$ pour $n=2k+1$  
Cette version n’est pas terminale.  

**Version 4**  
Ecrire la version terminale ```puissance_rec2_term```.


````python
# Version 1 #

def puissance_rec1(x, n):
    if n == 0:
        return 1
    else:
        return x * puissance_rec1(x, n - 1)
````

````python
puissance_rec1(3, 3)
````

````python
puissance_rec1(5, 2)
````

````python
puissance_rec1(10, 5)
````

````python
# Version 2 #
def puissance_rec1_term(x, n, acc = 1):
    if n == 0:
        return acc
    else:
        return puissance_rec1_term(x, n - 1, acc * x)
````

````python
puissance_rec1_term(5, 2)
````

````python
puissance_rec1_term(10, 5)
````

````python
puissance_rec1_term(7, 1)
````

````python
puissance_rec1_term(4, 0)
````

````python
# Version 3 #
def puissance_rec2(x, n):
    if n == 0:
        return 1
    else:
        if n % 2 == 0:
            return puissance_rec2(x, n // 2) **2
        else:
            return x * puissance_rec2(x, (n - 1) // 2) **2
````

````python
# Version 4 #
def puissance_rec2_term(x, n, acc = 1):
    if n == 0:
        return acc
    else:
        if n % 2 == 0:
            return puissance_rec2_term(x, n // 2, acc * x)
        else:
            return puissance_rec2_term(x*x, (n - 1) // 2, acc * x)
````
