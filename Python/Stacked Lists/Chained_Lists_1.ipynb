#   <div align="center">Structure linéaire</div>

Définition : Une structure linéaire sur un ensemble E est une suite d'éléments de E, c'est-à-dire une liste ordonnée d'éléments de E qui peut avoir deux formes :
1. les élements ont un indice qui permet d'accéder directement à cette place : il s'agit d'un tableau. L'avantage est la rapidité d'accés à un élément particulier.
2. chaque place a un successeur ou/et un prédécesseur et l'accés a cette place n'est plus direct : il s'agit d'une liste chaînée. L'avantage est le dynamisme de la structure.

## 2. Listes chaînée

L'élément de base d'une liste simplement chaînée est une cellule constituée de deux parties.
* La première contient une donnée - par exemple un entier pour une liste d'entiers.
* La seconde contient un pointeur (i.e. une adresse mémoire) vers une autre cellule (ou un pointeur nul).

Une liste est alors une succession de cellules, chacune pointant sur la suivante, et la dernière ayant un pointeur nul - i.e. ne pointant sur rien. En pratique, la variable «contenant» la liste est simplement un pointeur vers la première cellule. 

Voici par exemple une représentation d'une liste contenant les entiers 2, 4, 1 et 5.  
<img src="liste_chainee.png" width=500 height=500 />

En python, il n'existe pas de type correspondant aux listes chaînées. On peut implémenter cette structure à l'aide du type tuple ou d'une classe.

````python
class Maillon:
    def __init__(self, valeur = None, suivant = None):
        self.val = valeur
        self.suiv = suivant
        
    def __str__(self):
        if self.val is not None:
            return str(self.val) + " - " + str(self.suiv)
        else:
            return str(self.val)
````

````python
class Liste_Chaine:
    def __init__(self, premier = None):
        self.prem = Maillon(premier)
        
    def ajoute(self, valeur):  #ajout d'un élément en fin de liste
        m = Maillon(valeur)    # création d'un maillon
        if self.prem.val is None:  # si la liste est vide
            self.prem = m
        else:
            p = self.prem
            while p.suiv is not None:  # recherche du dernier élément
                p = p.suiv
            p.suiv = m  # m est le dernier élément
    
    def __str__(self):
        return str(self.prem)
````

On peut compléter le code de la classe ```Liste_Chaine``` avec une fonction ```tete``` qui renvoie la tête de la liste et une fonction ```queue``` renvoyant la liste obtenue après le retrait du 1er élément.

````python
class Liste_Chaine:
    def __init__(self, premier = None):
        self.prem = Maillon(premier)
        
    def ajoute(self, valeur):  #ajout d'un élément en fin de liste
        m = Maillon(valeur)    # création d'un maillon
        if self.prem.val is None:  # si la liste est vide
            self.prem = m
        else:
            p = self.prem
            while p.suiv is not None:  # recherche du dernier élément
                p = p.suiv
            p.suiv = m  # m est le dernier élément
    
    def __str__(self):
        return str(self.prem)
    
    def tete(self):
        if self.prem is not None:
            return self.prem.val
    
    def queue(self):
        if self.prem.val is None:
            return None
        else:
            liste = Liste_Chaine()
            liste.prem = self.prem.suiv
            return liste
````

````python
class Liste_Chaine:
    def __init__(self, premier = None):
        self.prem = Maillon(premier)
        
    def ajoute(self, valeur):  #ajout d'un élément en fin de liste
        m = Maillon(valeur)    # création d'un maillon
        if self.prem.val is None:  # si la liste est vide
            self.prem = m
        else:
            p = self.prem
            while p.suiv is not None:  # recherche du dernier élément
                p = p.suiv
            p.suiv = m  # m est le dernier élément
    
    def __str__(self):
        return str(self.prem)
    
    def tete(self):
        if self.prem is not None:
            return self.prem.val
    
    def queue(self):
        if self.prem.val is None:
            return None
        else:
            liste = Liste_Chaine()
            liste.prem = self.prem.suiv
            return liste
    
    def insere1(self, valeur):
        m = Maillon(valeur)
        if self.prem.val is None:  # si la liste est vide
            self.prem = m
        else:
            p = self.prem  # p est le 1er élément actuel
            self.prem = m  # m devient le 1er élément
            m.suiv = p  # p devient le suivant de m
        
````

Comparaison avec None : A priori, il n'y a pas de différence entre ```self.val is not None``` et ```self.val != None``` ou en positif ```self.val is None``` et ```self.val == None```. **Les deux opérations ```is``` et ```==``` ne sont pas les mêmes :**
* L'opération ```is``` est une égalité physique (être identique à un objet, au même endroit dans la mémoire).
* L'opération ```==``` est une égalité structurelle (être la même valeur, après une comparaison en profondeur).  

Dans le cas de la valeur ```None```, ces deux opérations coïncident, car l'objet que représente ```None``` est unique.
Mais dans une classe, on préferera l'utilisation de ```is```.
