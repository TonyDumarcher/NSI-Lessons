# Pile et POO
Les représentations précédentes utilisent un type existant, le type ```list```.
Dans la programmation orientée objet, on définir une structure nouvelle.  
Nous considérons des piles à capacité non bornée.  
On définit une classe ```Pile``` avec : 

des **attributs**:
* taille : un entier (nombre d'éléments contenus dans la pile);
* éléments (la pile) : une liste de longueur taille (type list); 

des **méthodes**:
* procédure ```init```, le constructeur (fixe la taille à 0 et crée une liste vide);
* procédure ```empiler```;
* fonction ```depiler```;
* fonction ```pile_vide```;
* fonction ```sommet```;
* des méthodes d'affichages.  

Ecrire le programme de cette nouvelle structure puis la tester.

````python
class Pile:
    def __init__(self, taille):
        self.taille = taille
        self.pile = []

    def empiler(self, element):
        if len(self.pile) < self.taille:
            self.pile.append(element)
        else:
            return None
        return self.pile
    
    def depiler(self):
        if not self.pile_vide():
            return self.pile.pop()
        return None
    
    def pile_vide(self):
        return len(self.pile) == 0
    
    def sommet(self):
        if not self.pile_vide():
            return self.pile[-1]

    def __str__(self):
        return str(self.pile)
````

````python
p1 = Pile(3)
Pile.empiler(p1, 1)
Pile.empiler(p1, 2)
Pile.empiler(p1, 3)
print(p1)
Pile.pile_vide(p1)
````

````python
Pile.sommet(p1)
````

````python
Pile.depiler(p1)
Pile.depiler(p1)
print(p1)
Pile.sommet(p1)
````
