    <h3>Liste chaînée à l'aide d'un tuple</h3></div>

On définit en Python une liste chaînée à l'aide d'un tuple :

````python
def liste_ch(*args):
#*args récupére l'ensemble des arguments donnés et les assemble sous forme de liste (packing).
    liste = None
    for val in reversed(args):  #inverse les l'ordre des éléments d'une liste sans modifier la liste
        liste = (val,liste)
    return liste



def tete(liste):
    if liste is not None:
        return liste[0]

def queue(liste):
    if tete(liste) is not None:
        return liste[1]
    
def ajoute(val, liste):
    return (val, liste)
    
def element_n(n, liste):
    element = liste
    numero = 0
    while element is not None:
        if numero == n:
            return tete(element)
        element = queue(element)
        numero += 1
    
def longueur(liste):
    if queue(liste) is None:
        return 1
    else:
        return 1 + longueur(queue(liste))
    
maliste = liste_ch(2, 8, 3, 5, 4)
print(tete(maliste))
print(queue(maliste))
print(ajoute(1, maliste))
print(element_n(1, maliste))
print(longueur(maliste))
````
