Ecrire les fonctions selon l'interface suivante qui permet de traiter les deux implémentations :  

| Fonctions 	| Description 	|
|-	| -	|
| vide(arbre) 	| renvoie ```True``` si l'arbre est une liste vide ou ```None```, sinon ```False``` 	|
| gauche(arbre) 	| renvoie ```arbre[1]``` ou ```arbre.gauche``` selon le type de ```arbre```. *On utilisera la fonction ```isinstance()```.*	|
| droit(arbre) 	| renvoie ```arbre[2]``` ou ```arbre.droit``` selon le type de ```arbre```. *On utilisera la fonction ```isinstance()```.*	 	|
|taille(arbre)	| renvoie 0 si l'arbre est vide, sinon la somme du noeud racine, du nombre des noeuds du sous-arbre gauche et du nombre des noeuds du sous-arbre droit. *On utilisera les trois fonctions précédentes.*	|
| hauteur(arbre) 	| renvoie 0 si l'arbre est vide, sinon la somme du noeud racine, le maximum entre le nombre de noeuds du sous-arbre gauche et le nombre de noeuds du sous-arbre droit.   *On utilisera de même les trois premières fonctions.*	|
| affiche_valeur(arbre) 	| affiche la valeur du sommet de l'arbre. *On utilisera la fonction ```isinstance()```.*  |

````python
def vide(arbre):
  return True if arbre == [] or arbre is None else False

def gauche(arbre):
  return arbre[1] if isinstance(arbre, list) else arbre.gauche

def droit(arbre):
  return arbre[2] if isinstance(arbre, list) else arbre.droit

def taille(arbre):
  return 0 if vide(arbre) else (1 + taille(gauche(arbre)) + taille(droit(arbre)))

def hauteur(arbre):
  return 0 if vide(arbre) else (1 + max(hauteur(gauche(arbre))), hauteur(droit(arbre)))

def affiche_valeur(arbre):
  print(arbre[0] if isinstance(arbre, list) else arbre.valeur)
````

**Exercice n°2**
Programmer en Python les trois algorithmes donnés au TD1 qui permettent de parcourir un arbre en profondeur selon l'implémentation suivante :  

| Fonctions 	| Description 	|
|-	|:-	|
| dfs_prefixe(arbre) 	| parcourt en profondeur l'arbre dans l'ordre préfixe et affiche les noeuds dans l'ordre parcouru. |
| dfs_infixe(arbre) 	| parcourt en profondeur l'arbre dans l'ordre infixe et affiche les noeuds dans l'ordre parcouru. |
| dfs_postfixe(arbre) 	| parcourt en profondeur l'arbre dans l'ordre postfixe et affiche les noeuds dans l'ordre parcouru. |

````python
# Parcours Préfixe

def dfs_prefixe(arbre):
  if not vide(arbre):
    affiche_valeur(arbre)
    dfs_prefixe(gauche(arbre))
    dfs_prefixe(droit(arbre))

# Parcours Infixe

def dfs_infixe(arbre):
  if not vide(arbre):
    dfs_infixe(gauche(arbre))
    affiche_valeur(arbre)
    dfs_infixe(droit(arbre))

# Parcours Postfixe

def dfs_postfixe(arbre):
  if not vide(arbre):
    dfs_postfixe(gauche(arbre))
    dfs_postfixe(droit(arbre))
    affiche_valeur(arbre)
````
