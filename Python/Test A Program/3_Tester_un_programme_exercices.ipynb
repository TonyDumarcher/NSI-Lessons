Ce programme contient une erreur qui correspond à la catégorie des erreurs (bugs) permanents et non visibles.  
Tester et corriger cette erreur.  
On suppose que la liste en paramètre contient au moins cinq éléments.

````python
def lissage(liste):
    n = len(liste)
    lis = [liste[0],liste[0]]
    for i in range(2,n-2):
        lis.append(sum(liste[i-2:i+2])/5)
    lis.append(liste[n-2])
    lis.append(liste[n-1])
    return lis
````

````python
def lissage(liste):
    try:
        if not isinstance(liste, list):
            raise TypeError("L'entrée doit être une liste.")
        
        n = len(liste)
        if n < 5:
            raise ValueError("La liste doit contenir au moins 5 éléments.")

        lis = [liste[0], liste[1]]
        for i in range(2, n - 2):
            # i+3 pour inclure le 5ème élément
            moyenne = sum(liste[i - 2 : i + 3]) / 5
            lis.append(moyenne)
            
        lis.append(liste[n-2])
        lis.append(liste[n-1])
        return lis

    except (TypeError, ValueError) as error:
        print(f"Erreur : {error}")
        return None
````

**Exercice n°6**  
Le programme qui suit contient deux fonctions ```double1``` et ```double2```. Ces fonctions prennent une liste en paramètre et leur rôle doit être de concaténer la liste avec elle-même. Si la liste vaut ```[1, 2]``` après l'exécution d'une fonction la liste doit avoir pour valeur ```[1, 2, 1, 2]```.

````python
def double1(liste):
    for i in range(len(liste)):
        liste.append(liste[i])
        
def double2(liste):
    i = 0
    while i < len(liste):
        liste.append(liste[i])
        i = i + 1
````

1. L'une des deux fonctions ne remplit pas son rôle et ne termine pas. Expliquer la non-terminaison.  
2. Modifier le code de la fonction qui ne termine pas en gardant le même style de boucle afin qu'elle remplisse son rôle correctement.

La fonction double2 utilise une boucle "while i < len(liste):" qui est infinie vue que le i est incrémenté à chaque répétition, la longueur le la liste augmente aussi, résultat : i sera toujours inférieur à la longueur de la liste

````python
def double2(liste):
    i = 0
    n = len(liste)
    while i < n:
        liste.append(liste[i])
        i = i + 1
````
