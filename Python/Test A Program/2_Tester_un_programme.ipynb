**Programmation défensive** : Si on veut éviter qu'une utilisation de cette fonction sur un tableau vide produise un message d'erreur, on peut interrompre le programme dès que la fonction reçoit un tableau vide.
1. on peut tester la taille du tableau à l'entrée et utiliser la fonction ```exit``` qui interrompt le programme définitivement, après avoir affiché le message passé en argument.

````python
import sys
def mystere(t):
    if len(t) == 0:
        sys.exit("Le tableau est vide")
    m = 0
    for i in range(len(t)):
        if t[i]>t[m]:
            m = i
    return m
````

````python
t=[]
print(mystere(t))
````

2. Une meilleure solution consiste à utiliser ```assert``` qui combine le test d'une condition et l'interruption d'un programme avec un message dans le cas où cette condition n'est pas validée.

````python
def mystere(t):
    assert len(t) > 0, "Le tableau est vide"
    m = 0
    for i in range(len(t)):
        if t[i]>t[m]:
            m = i
    return m
````

````python
t=[]
mystere(t)
````

On appelle cela de la *programmation défensive*.

**Tester un programme**  
Plutôt que d'effectuer des tests les uns après les autres, on peut les inclure dans le programme. Par exemple, avec la construction ```assert``` :

````python
def mystere(t):
    if len(t) == 0:
        return None
    m = 0
    for i in range(len(t)):
        if t[i]>t[m]:
            m = i
    return m
````

````python
assert mystere([2, 3, 1]) == 1
assert mystere([]) == None
assert mystere([3, 1, 3, 7]) == 3
assert mystere([8, 3, 1, 3, 7]) == 0
assert mystere([-3, -1, -3, -7]) == 1
t=[1, 2, 3]
print(mystere(t))
````

Si l'un des tests échoue, il faut rectifier le programme. Une fois la correction faite, il faut relancer les tests, y compris ceux qui avaient eu succès.

**Exceptions**  
L'erreur vue lors du 1er test, intitulée ```IndexError``` est appelée une **exception**. De telles erreurs correspondent à la détection par l'interprète Python d'un problème empêchant la bonne exécution du programme. Lorsque l'exception intervient le programme est interrompu sur le champ.  
Quelques exceptions courantes :
* NameError  : accès à une variable inexistante.
* IndexError  : accès à un indice invalide d'un tableau.
* KeyError : accès à une clé inexistante d'un dictionnaire.
* ZeroDivisionError : division par zéro.
* TypeError : opération appliquée à des valeurs incompatibles.  

Il est possible de déclencher directement toutes ces exceptions (on dit "lever une exception") avec l'opérateur ```raise``` de Python.

**Rattraper une exception**  
Les exceptions levés dans un programme peuvent être causées par des erreurs du programme : elles sont imprévues et leurs conséquences ne sont, par définition, pas maîtrisées. Par exemple, si on entre le mot "erreur" au lieu d'un nombre dans la demande suivante : 

````python
nombre = int(input("Entrer un nombre : "))
````

Dans ce cas, la fonction ```int``` lève une exception ```
ValueError```.  
cependant, en tant que programmeur, on peut anticiper cette éventualité et prévoir un comportement alternatif du programme pour cette situation exceptionnelle. Il faut alors "rattraper" cette exception, c'est-à-dire l'intercepter avant que l'exécution du programme ne soit définitivement abandonné, avec la construction suivante : 

````python
try:
    nombre = int(input("Entrer un nombre : "))
except ValueError:
    print("Prière d'entrer uniquement un nombre entier")
````

Le mot-clé ```try``` suivi du symbole ```:``` introduit un 1er bloc de code, puis le mot-clé ```except``` suivi du nom de l'exception et du symbole ```:``` précède un 2ème bloc de code. On qualifie le 1er bloc de *normal* et le 2ème d'*alternatif*. Si à l'exécution, le bloc normal ne lève pas d'exception, alors le bloc alternatif est ignoré et on passe aux instructions suivantes. Si, à l'inverse, une exception est levée, alors l'exécution du bloc normal est immédiatement interrompue et le nom de l'exception levée est comparée avec le nom précisé à la ligne ```except```. Si les noms correspondent, l'exception est *rattrapée* et on exécute le bloc alternatif avant, éventuellement, de passer à la suite. Sinon, l'exception est *propagée* et le programme s'interrompt, à moins que le fragment de ce code soit lui-même inclus dans une autre construction ```try/except``` qui pourra à son tour tenter de rattraper l'exception. Par exemple : 

````python
somme = 0
while True:
    try:
        nombre = int(input("Entrer un nombre : "))
        somme += nombre
        break
    except ValueError:
        print("Prière d'entrer uniquement un nombre entier")
````

L'instruction ```break``` met fin à la boucle infinie.  

**Remarques** :  
1. Si plusieurs exceptions peuvent être rattrapées, il est possible d'écrire plusieurs blocs alternatifs, chacun associé à sa ligne ```except```. On parle alors d'**alternatives multiples**.  
Par exemple :  
```python
try:
    <bloc normal>
except Exception 1:
    <bloc alternatif 1>
except Exception 2:
    <bloc alternatif 2>
```

2. Si on ne précise pas le nom d'exception après ```except```, alors *toutes* les exceptions seront rattrapées par ce bloc alternatif :  
```python
try:
    quelque_chose_de_dangeureux()
except:
    erreur_connue_mais_aussi_erreur_inconnue
```
Par ce code, on risque d'effacer d'eventuelles erreurs non prévues ou connues ce qui complique le diagnostic.  

Autre **exemple** avec le module ``` Turtle``` :  
En utilisant ce module dans une feuille Jupyter version Bureau, Jupyter exécute du code Python dans une page web. Turtle crée une interface graphique bureau (gérée par le système d’exploitation local et Tkinter). La fenêtre Turtle est mal intégrée dans le navigateur, ce qui pose deux difficultés :
* Un problème de fermeture de la fenêtre
* La levée d’une exception de type Terminator au lancement d’un code Turtle (en fait une fois sur deux).

````python
from turtle import *

for _ in range(3):
    forward(80)
    up()
    left(120)
    forward(40)
    down()
    left(120)

exitonclick()
````

Pour y remédier, on utilisera le code boilerplate, un code passe-partout, qui lève l'exception :  

````python
from turtle import *

try:
    reset()
except Terminator:
    pass

for _ in range(3):
    forward(80)
    up()
    left(120)
    forward(40)
    down()
    left(120)

exitonclick()
````

Ainsi, le programme peut être exécuté deux fois sans lever d'erreur.
