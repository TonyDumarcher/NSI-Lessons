**Exercice POO**  
Dans certains langages de programmation, comme Pascal ou Ada, les tableaux ne sont pas nécessairement indexés à partir de 0. C'est le programmeur qui choisit sa plage d'indices. Par exemple, on peut déclarer un tableau dont les indices vont de -10 à 9 si on le souhaite. Dans cet exercice, on se propose de construire une classe ```Tableau``` pour réaliser de tels tableaux. Un objet de cette classe aura deux attributs, un attribut ```premier``` qui est la valeur du premier indice et un attribut ```contenu``` qui est un tableau Python contenant les élements. Ce dernier est un vrai tableau Python, indexé à partir de 0.
* Ecrire un constructeur ```__init__(self, imin, imax, v) ``` où ```imin``` est le premier indice, ```imax``` le dernier indice et ```v``` la valeur utilisée pour initialiser toutes les cases du tableau. Ainsi, on peut écrire ``` t=Tableau(-10,9,42)``` pour construire un tableau de vingt cases indexées de -10 à 9 et toutes initialisées à 42.
* Ecrire une méthode ```__len__(self)``` qui renvoie la taille du tableau.
* Ecrire une méthode ```__getitem__(self, i)``` qui renvoie l'élement du tableau ``` self``` d'indice ```i```. De même, écrire une méthode ```__setitem__(self, i, v)``` qui modifie l'élement du tableau ``` self``` d'indice ```i``` pour lui donner la valeur ``` v```. Ces deux méthodes doivent vérifier que l'indice ```i``` est bien valide et, dans le cas contraire, lever l'exception ```IndexError``` avec la valeur de ```i```  en argument (c'est-à-dire ```raise IndexError(i)```). Pour cela, créer une méthode ```_indice_valide(self,i)```.
* Enfin, écrire une méthode ```__str__(self)``` qui renvoie une chaîne de caractères décrivant le contenu du tableau.

````python
class Tableau():
    def __init__(self, imin, imax, v):
        if imin > imax:
            raise ValueError
        else:
            self.premier = imin
            self.dernier = imax
            self.v = v
            self.contenu = (imax - imin + 1) * [self.v]

    def __len__(self):
         return len(self.contenu)

    def __getitem__(self, i):
        if i < self.premier or i > self.dernier:
            raise IndexError
        else:
            return self.contenu[self.premier + i]

    def __setitem__(self, i, v):
        if Tableau._indice_valide(self, i):
            self.contenu[self.premier + i] = v
            return self.contenu[self.premier + i]
        
    def _indice_valide(self, i):
        if self.premier < i < self.dernier:
            return True
        else:
            return IndexError

    def __str__(self):
        return str(self.contenu)
````

````python
### TESTS CREES PAR L'IA ###

print("--- DÉBUT DES TESTS DE LA CLASSE TABLEAU ---")

## 1. Tests de création et de longueur

print("\n## 1. Création et Longueur")

# Test 1: Indices standard (0 à N)
t1 = Tableau(0, 4, 'A')
print(f"T1 (0 à 4, init 'A') : {t1}")
print(f"T1 Longueur : {len(t1)}") # Attendu: 5

# Test 2: Indices négatifs et non nuls
t2 = Tableau(-5, -2, 100)
print(f"T2 (-5 à -2, init 100) : {t2}")
print(f"T2 Longueur : {len(t2)}") # Attendu: 4

# Test 3: Indices chevauchant zéro
t3 = Tableau(-2, 2, 0)
print(f"T3 (-2 à 2, init 0) : {t3}")
print(f"T3 Longueur : {len(t3)}") # Attendu: 5

# Test 4: Tableau de taille 1 (cas limite)
t4 = Tableau(10, 10, True)
print(f"T4 (10 à 10, init True) : {t4}")
print(f"T4 Longueur : {len(t4)}") # Attendu: 1


## 2. Tests d'accès et de modification

print("\n## 2. Accès et Modification")

# Test 5: Accès aux bornes et modification
t5 = Tableau(-3, 3, 5) # [-3: 5, -2: 5, -1: 5, 0: 5, 1: 5, 2: 5, 3: 5]
print(f"T5 Initial: {t5}")
t5[-3] = "PREMIER" # Modification de l'indice minimum
t5[3] = "DERNIER" # Modification de l'indice maximum
t5[0] = "ZERO"    # Modification de l'indice 0
t5[-1] = 42       # Modification d'un indice négatif
print(f"T5 Modifié: {t5}")
print(f"T5 Élément à l'indice -3 : {t5[-3]}") # Attendu: PREMIER
print(f"T5 Élément à l'indice 3 : {t5[3]}")   # Attendu: DERNIER
print(f"T5 Élément à l'indice -1 : {t5[-1]}") # Attendu: 42


## 3. Tests des cas d'erreur (IndexError)

print("\n## 3. Gestion des erreurs (IndexError)")

# Test 6: Accès en dehors de la borne inférieure (trop petit)
t6 = Tableau(1, 5, 0) # Plage valide [1, 5]
try:
    val = t6[0]  # Indice 0 est invalide
    print(f"Test 6 - ÉCHEC: L'accès à t6[0] aurait dû lever IndexError. Valeur obtenue: {val}")
except IndexError as e:
    print(f"Test 6 - OK: IndexError levée pour l'indice en dehors de la borne inférieure. Erreur: {e}")

# Test 7: Modification en dehors de la borne supérieure (trop grand)
try:
    t6[6] = 99  # Indice 6 est invalide
    print("Test 7 - ÉCHEC: La modification t6[6] aurait dû lever IndexError.")
except IndexError as e:
    print(f"Test 7 - OK: IndexError levée pour l'indice en dehors de la borne supérieure. Erreur: {e}")

# Test 8: Accès avec indices négatifs (trop petit)
t8 = Tableau(-4, -1, 0) # Plage valide [-4, -1]
try:
    val = t8[-5]  # Indice -5 est invalide
    print(f"Test 8 - ÉCHEC: L'accès à t8[-5] aurait dû lever IndexError. Valeur obtenue: {val}")
except IndexError as e:
    print(f"Test 8 - OK: IndexError levée pour l'indice négatif trop petit. Erreur: {e}")
    
# Test 9: Tentative de création avec imin > imax
try:
    t9 = Tableau(5, 1, 0)
    print("Test 9 - ÉCHEC: La création aurait dû lever ValueError.")
except ValueError as e:
    print(f"Test 9 - OK: ValueError levée pour imin > imax. Erreur: {e}")
    
print("\n--- FIN DES TESTS ---")
````
