**Bibliothèque**   
Sous windows ou linux, on installe une bibliothèque ou un module en téléchargeant un ou plusieurs packages, c'est-à-dire un ensemble de dossiers et sous-dossiers. Ce package contient un fichier ```__init__.py``` pour indiquer à python qu'il ne s'agit pas d'un simple répertoire mais bien d'un package.  
**Exemple** : après l'installation de la bibliothèque Matplolib, un dossier matplolib est créé et contient divers modules comme ```pylab``` ou ```pyplot```. Ces modules sont constitués des fichiers ```pylab.py``` et ```pyplot.py```. pour utiliser le module ```pyplot```, on écrit l'instruction ```import matplolib.pyplot```.
Pour exécuter les programmes suivants, on écrit ```import matplotlib.pyplot as plt```. Ce qui signifie que pour utiliser une fonction de ce module comme ```show()``` par exemple, on devra écrire ```plt.show()``` (puisqu'on a importé le module sous le nom plt).
De plus, le module matplotlib est très lié à un autre module qui sert à faire du calcul numérique qui s'appelle numpy et qu'on importe souvent sous le nom np.

Pour résumer, pour représenter graphiquement des fonctions ou autres, il faudra mettre en en-tête :

````python
import matplotlib.pyplot as plt
import numpy as np
````

```plt.show()``` permet d'afficher le résultat.  
**Quelques programmes**

````python
plt.plot([1,3,4],[2,1,6])

plt.show()
````

Décrire la fonction ```plt.plot```

````python
x = np.linspace(-2, 2, 100)
y = 2*x**2+3*x-4

plt.plot(x,y)
plt.show()
````

Décrire la fonction ```np.linspace```.

````python
abscisses = np.linspace(-4,4,100)
ordonnées = [np.cos(x)+3*np.sin(2*x) for x in abscisses]
plt.plot(abscisses,ordonnées)
plt.show()
````

**Autre exemple**

````python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-np.pi,np.pi,100)
for n in range(1,5):
    y = np.cos(n*x)
    plt.plot(x,y, label="n="+str(n))
    
plt.legend(loc="lower right")
plt.show()
````

La fonction ```plt.legend()``` permet de rajouter une légende pour que le graphique soit plus compréhensible. On précise sa position (ici en bas à droite).  
On remarquera que nous avons utilisé ```np.pi```, la constante pi étant disponible dans le module numpy.  
En important le module ```math```, on aurait pu écrire le programme suivant :

````python
import matplotlib.pyplot as plt
import numpy as np
from math import *

x = np.linspace(-pi,pi,100)
for n in range(1,5):
    y = np.cos(n*x)
    plt.plot(x,y, label="n="+str(n))
    
plt.legend(loc="lower right")
plt.show()
````

**Exercice 4**  
En écrivant les modules importés, afficher les représentations graphiques de la famille de fonctions $y =$ ${1−nx} \over {x−1}$ pour $n$ allant de $-3$ à $3$.  
On se placera dans un repère allant de $-2$ à $4$ pour les abscisses et de $-8$ à $8$ pour les ordonnées. On fera apparaitre aussi une légende en haut à gauche.

**API WEB**  
Pour ce type d'API, les demandes ne sont pas codés dans le même langage que le code de l'API. Une méthode est d'émettre une requête vers un serveur distant qui renvoie alors les données demandées.  
Pour effectuer des requêtes en python, le plus simple est d'utiliser la bibliothèque ```requests```. 

````python
import requests
````

Afin de récupérer des données, il faut émettre une requête de type ```get```.

````python
rep = requests.get("https://www.education.gouv.fr/test")
rep
````

````python
rep.status_code
````

Le code d'erreur 404 est renvoyé, ce qui signifie que la page n'existe pas.  
Les différents types de code réponses à une requête ```get``` sont :  
* 200 : tout est ok, le résultat est renvoyé;
* 301 : le serveur est redirigé vers une autre adresse;
* 400 : la requête ne semble pas correcte pour le serveur;
* 401 : il y a un problème d'authentification;
* 403 : l'accès à la ressource n'est pas autorisée;
* 404 : la ressource n'a pas été trouvée.

Voici une requête valide pour obtenir le nom, le code et la population des communes du département 06 : 

````python
url = """https://geo.api.gouv.fr/departements/06/communes?fields=nom,code,population&format=json"""
rep = requests.get(url)
rep.status_code
````

On reconnaît dans l'URL la forme d'une requête ```get```.  
Les données renvoyées sont au format ```JSON```. Pour les afficher, la méthode ```json()``` produit une liste.  
On demande d'afficher les 4 premiers éléments :

````python
rep.json()[:4]
````

On peut recherche des communes par nom, par code postal ou par coordonnées géographiques.  
Par code postal : 

````python
url = "https://geo.api.gouv.fr/communes?codePostal=06600"
rep = requests.get(url)
rep.json()
````

Une autre API permet la recherche d'adresse : https://geo.api.gouv.fr/adresses.

````python
url = """https://api-adresse.data.gouv.fr/search/?q=Boulevard+de+Bacon&limit=4"""
rep = requests.get(url)
rep.json()['features'][0]['properties']
````

Un dernier exemple d'utilisation d'une API avec GéoDataMine https://geodatamine.fr/ qui permet d'extraire des données de OpenStreetMap France https://www.openstreetmap.fr/

````python
url = """https://geodatamine.fr/boundaries"""
r = requests.get(url)
data=r.json()
for d in data:
    k=d["name"]
    if k == "Antibes":
        print(d)
        break #permet d'écrire l'identifiant de la ville d'Antibes 
              #en une seul ligne
````

On utilise l'identifiant obtenu pour obtenir la liste des bibliothèques (de livres) de la ville :

````python
url = """https://geodatamine.fr/data/library/-74687?format=geojson"""
r = requests.get(url)
print(r.json())
````
