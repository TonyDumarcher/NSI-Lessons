**Exercice 8**  

Écrire une fonction deconvers(mot,b) qui fait le contraire de la fonction de l'exercice 7.

Aide : l'instruction print(signes.index("A")) affiche 10.

Par exemple : Si mot = "FA3 " , signes.index(mot[1]) renvoie 10

````python
""" Exercice 8 """

def deconvers(mot, b):
    signes = "0123456789ABCDEFGHIJKLMNOPQRSTUVWXYZ"
    n = 0
    for character in mot:
        n = n * b + signes.index(character)
    return n

print(deconvers("FA3", 16))
print(deconvers("1010", 2))
````

**Exercice 9**  

En effectuant vos propres recherche, écrire une fonction qui affiche les n premiers termes de la suite de Fibonacci.

````python
""" Exercice 9 """

def suite_fibonacci(n):
    fibonacci = 0
    suite = [1, 1]
    if n == 1:
        return "[1]"
    for i in range(1, n-1):
        fibonacci = suite[i - 1] + suite[i]
        suite.append(fibonacci)
    return suite

print(suite_fibonacci(1))
print(suite_fibonacci(2))
print(suite_fibonacci(3))
print(suite_fibonacci(5))
print(suite_fibonacci(10))
print(suite_fibonacci(20))
````

**Exercice 10**  

En vous inspirant des méthodes expliquées sur cette page : https://www.jeuxmaths.fr/cours/calcul-PGCD.php, réaliser une fonction qui renvoie le PGCD de deux entiers naturels.

````python
""" Exercice 10 """

def PGCD(n1, n2):
    while n1 > 0 or n2 > 0:
        var_temporaire = 0
        if n1 > n2:
            var_temporaire = n1 - n2
            n1 = var_temporaire
        else:
            var_temporaire = n2 - n1
            n2 = var_temporaire
        if n1 == 0:
            return n2
        elif n2 == 0:
            return n1

PGCD(351, 208)
````
