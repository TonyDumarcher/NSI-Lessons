```python
def ISBN(code):
    if len(code) != 13:
        return False
    total = 0
    for i in range(12):
        total += int(code[i])
    for i in range(1, 12, 2):
        total += int(code[i]) * 2
    cle_calculee = 10 - (total % 10)
    if cle_calculee == int(code[12]):
        return True
    else:
        return False

print(ISBN("9782729877828"))
```

# Exercice n°6

```python
def initiales(nom):
    majuscules = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
    initiales = ''
    for lettre in nom:
        if lettre in majuscules:
            initiales = initiales + lettre
    return initiales

print(initiales("Vincent Latour"))
print(initiales("Jean-Jacques Goldman"))
print(initiales("Jean-Claude Van Damme"))
```

```python
def nb_mots(phrase):
    nb_mots = 1     # Il y a toujours au moins 1 mot dans une phrase
    for lettre in phrase:
        if lettre == ' ':
            nb_mots = nb_mots + 1
    return nb_mots

nb_mots('Dans le jardin, les fleurs colorées attirent de nombreux papillons et oiseaux chanteurs qui volent autour des plantes.')
```

# Exercice n°8

```python
def lettre_maxi(texte):
    lmax = 0
    texte = texte.lower()
    lettremax = ""
    alphabet = 'abcdefghijklmnopqrstuvwxyz'
    for lettre in alphabet:
        lettre_temp = 0
        for character in texte:
            if lettre == character:
                lettre_temp = lettre_temp + 1
        if lettre_temp > lmax:
            lmax = lettre_temp
            lettremax = lettre
    return lmax, lettremax
    
lettre_maxi("Les enfants rient et courent dans le parc pendant l'après-midi ensoleillé.")
```

# Exercice n°11

```python
def full(main):
    compte = {}
    for carte in main:
        if carte in compte:
            compte[carte] += 1
        else:
            compte[carte] = 1
    brelan = False
    paire = False
    for cle in compte:
        if compte[cle] == 3:
            brelan = True
        elif compte[cle] == 2:
            paire = True
    return brelan and paire

main_exemple = ["As", "As", "As", "Roi", "Roi"]
print(full(main_exemple))
```

# Exercice n°12

```python
def mastermind(proposition, reponse):
    bien_places = 0
    reste_prop = []
    reste_rep = []
    for i in range(len(proposition)):
        if proposition[i] == reponse[i]:
            bien_places += 1
        else:
            reste_prop.append(proposition[i])
            reste_rep.append(reponse[i])
    mal_places = 0
    for p in reste_prop:
        if p in reste_rep:
            mal_places += 1
            reste_rep.remove(p)
    return "bien placées:", bien_places, "mal placées:", mal_places

print(mastermind([1, 1, 2, 1, 3], [1, 7, 3, 2, 1]))
```
