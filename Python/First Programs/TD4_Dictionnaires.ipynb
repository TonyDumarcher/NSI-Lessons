```python
def separe_pnt(pnt):
    prenom = ""
    nom = ""
    tel = ""
    p = True
    n = False
    t = False
    for char in pnt:
        if p == True:
            if char != " ":
                prenom += char
            else:
                p = False
                n = True
        elif n ==  True:
            if char != " ":
                nom += char
            else:
                n = False
                t = True
        elif t == True:
            if char != " ":
                tel += char
    return prenom, nom, tel
```

## Exercice 37

```python
def cle(dict):
    for cle in dict:
        print(cle)

def valeurs(dict):
    for valeurs in dict:
        print(dict[valeurs])


def elements(dict):
    for cle_valeurs in dict:
        print(cle_valeurs)
        print(dict[cle_valeurs])
```

```python
capitales = {"Paris": "France", "Rome": "Italie", "Madrid": "Espagne"}

print("Cles:")
cle(capitales)
print("Valeurs:")
valeurs(capitales)
print("Elements:")
elements(capitales)
```
